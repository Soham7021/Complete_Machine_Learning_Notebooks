### Now We will create an empty data frame and append the data to it by iterating

```python
df = pd.DataFrame()
```

```python
df
```

```python
for i in range(1,249):
    response = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=1681aac8e7b08013e541788b79791a5c&language=en-US&page={}".format(i))
    # Here in above line we have modified the link at lastt of the link look at it carefully
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = df.append(temp_df,ignore_index=True)
```

```python
df.to_csv('movies.csv')
```
